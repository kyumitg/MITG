```python
dataCount = 20
fileFullPath = "./ECG/ECG-ID Database/rec_"

sig = np.array(np.zeros(dataCount),dtype=object)
fields = np.array(np.zeros(dataCount),dtype=object)
ann = np.array(np.zeros(dataCount),dtype=object)

for index in range(dataCount):
    sig[index], fields[index] = wfdb.rdsamp(fileFullPath + str(index+1))
    ann[index] = wfdb.rdann(fileFullPath + str(index+1), 'atr')

FilteredSig = np.array(np.zeros(dataCount),dtype=object)

for i in range(dataCount):
    size = sig[i][:,1].shape[0]
    FilteredSig[i] = np.array(np.zeros(size))
    FilteredSig[i] = sig[i][:,1]
```

```python
reValue = FilteredSig[0]*FilteredSig[0]*FilteredSig[0]

indexArray = np.array(np.zeros(100),dtype=int)

index = 0
for i in range(reValue.shape[0]):
    if(reValue[i] > 0.1
      and (reValue[i] >= reValue[i+1]) and (reValue[i] >= reValue[i-1])):
        indexArray[index]=i
        index+=1
        if(reValue[i] == reValue[i-1]):
            index-=1

print("Total Peek : " + str(index))
reShapeIndexArray = indexArray[:index]
#for i in reShapeIndexArray:
#    print(str(i) + " => " + str(reValue[i]))
```

```python

def LoadData(dataCount,fileFullPath):
    sig = np.array(np.zeros(dataCount),dtype=object)
    fields = np.array(np.zeros(dataCount),dtype=object)
    ann = np.array(np.zeros(dataCount),dtype=object)

    for index in range(dataCount):
        sig[index], fields[index] = wfdb.rdsamp(fileFullPath + str(index+1))
        ann[index] = wfdb.rdann(fileFullPath + str(index+1), 'atr')

    FilteredSig = np.array(np.zeros(dataCount),dtype=object)

    for i in range(dataCount):
        size = sig[i][:,1].shape[0]
        FilteredSig[i] = np.array(np.zeros(size))
        FilteredSig[i] = sig[i][:,1]
        
    return sig,fields,ann,FilteredSig

def GetSingleFilePeek(FilteredSig):
    reValue = FilteredSig*FilteredSig*FilteredSig

    indexArray = np.array(np.zeros(100),dtype=int)

    index = 0
    for i in range(reValue.shape[0]):
        if(reValue[i] > 0.1
          and (reValue[i] >= reValue[i+1]) and (reValue[i] >= reValue[i-1])):
            indexArray[index]=i
            index+=1
            if(reValue[i] == reValue[i-1]):
                index-=1

    reShapeIndexArray = indexArray[:index]
    return index,reShapeIndexArray
```

```python
dataCount = 20
fileFullPath = "./ECG/ECG-ID Database/rec_"

sig,fields,ann,FileteredSig = LoadData(dataCount,fileFullPath)

nowFileIndex = 0
IndexCount0, IndexArray0 = GetSingleFilePeek(FileteredSig[nowFileIndex])
print(IndexCount0)
for peek in IndexArray0:
    plt.plot(FilteredSig[nowFileIndex][peek-100:peek+200])
    plt.show()
```
